```python
# -*- coding: utf-8 -*-
"""
ã€Šæ´¥é—¨æ‚è®°ã€‹æ–‡æœ¬åˆ†æä»£ç  - åŒ…å«ä¸­æ–‡å­—ä½“è®¾ç½®
åˆ†æ19ä¸–çºªæœ«è‡³20ä¸–çºªåˆå¤©æ´¥å•†ä¸šç©ºé—´ä¸äº¤é€šæ–¹å¼å˜è¿
"""

# ========== 1. å®‰è£…å¿…è¦åº“å¹¶è®¾ç½®ä¸­æ–‡å­—ä½“ ==========
print("æ­£åœ¨è®¾ç½®ä¸­æ–‡å­—ä½“...")

# å®‰è£…å­—ä½“åº“ï¼ˆColabç¯å¢ƒï¼‰
!apt-get -qq install -y fonts-wqy-microhei
!apt-get -qq install -y fonts-wqy-zenhei

# å¯¼å…¥åŸºç¡€åº“
import pandas as pd
import numpy as np
import jieba
import re
import os
import warnings
from collections import Counter
warnings.filterwarnings('ignore')

# å®‰è£…å’Œå¯¼å…¥å¯è§†åŒ–åº“
print("å®‰è£…å¯è§†åŒ–åº“...")
!pip -q install matplotlib seaborn wordcloud

import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# è®¾ç½®ä¸­æ–‡å­—ä½“
import matplotlib
from matplotlib import font_manager

# æ·»åŠ ä¸­æ–‡å­—ä½“è·¯å¾„
font_dirs = ['/usr/share/fonts/truetype/wqy/']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)

for font_file in font_files:
    font_manager.fontManager.addfont(font_file)

# è·å–å¯ç”¨çš„ä¸­æ–‡å­—ä½“
chinese_fonts = ['WenQuanYi Micro Hei', 'WenQuanYi Zen Hei', 'DejaVu Sans',
                 'Arial Unicode MS', 'SimHei', 'Microsoft YaHei']

# è®¾ç½®matplotlibé»˜è®¤å­—ä½“
matplotlib.font_manager.fontManager.addfont("/usr/share/fonts/truetype/wqy/wqy-microhei.ttc")
matplotlib.font_manager.fontManager.addfont("/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc")

# è®¾ç½®å­—ä½“å‚æ•°
plt.rcParams['font.sans-serif'] = ['WenQuanYi Micro Hei', 'DejaVu Sans', 'sans-serif']
plt.rcParams['axes.unicode_minus'] = False  # è§£å†³è´Ÿå·æ˜¾ç¤ºé—®é¢˜

# éªŒè¯å­—ä½“æ˜¯å¦è®¾ç½®æˆåŠŸ
fig, ax = plt.subplots(figsize=(8, 4))
test_text = "ä¸­æ–‡å­—ä½“æµ‹è¯• - å¤©æ´¥å•†ä¸šäº¤é€š"
ax.text(0.5, 0.5, test_text, fontsize=20, ha='center', va='center')
ax.set_title("å­—ä½“æµ‹è¯•")
ax.axis('off')
plt.savefig('font_test.png', dpi=100, bbox_inches='tight')
plt.close()

print("ä¸­æ–‡å­—ä½“è®¾ç½®å®Œæˆï¼")

# ========== 2. æ•°æ®è¯»å–å‡½æ•° ==========
def read_text_file(filename):
    """
    è¯»å–æ–‡æœ¬æ–‡ä»¶ï¼Œå¤„ç†ç¼–ç é—®é¢˜
    """
    encodings = ['utf-8', 'gbk', 'gb2312', 'gb18030', 'utf-8-sig']
    for encoding in encodings:
        try:
            with open(filename, 'r', encoding=encoding) as f:
                text = f.read()
            print(f"âœ… æ–‡ä»¶ '{filename}' è¯»å–æˆåŠŸï¼Œç¼–ç : {encoding}")
            return text
        except UnicodeDecodeError:
            continue
    raise ValueError(f"âŒ æ— æ³•è¯»å–æ–‡ä»¶ {filename}ï¼Œè¯·æ£€æŸ¥ç¼–ç æ ¼å¼")

# ========== 3. æ–‡æœ¬é¢„å¤„ç†å‡½æ•° ==========
def load_stopwords(filename='stopwords.txt'):
    """
    åŠ è½½åœç”¨è¯åˆ—è¡¨
    """
    stopwords = set()

    # åˆ›å»ºé»˜è®¤åœç”¨è¯åˆ—è¡¨
    default_stopwords = {
        # æ–‡è¨€è™šè¯
        'ä¹‹', 'ä¹', 'è€…', 'ä¹Ÿ', 'çŸ£', 'ç„‰', 'å“‰', 'æ›°', 'äº‘', 'è€Œ', 'ä»¥',
        'å…¶', 'æ‰€', 'ä¸º', 'äº', 'åˆ™', 'ä½†', 'çš†', 'äº¦', 'åˆ', 'æœ‰', 'æ— ',
        'æœª', 'é', 'ä¸', 'ä¸”', 'ä¹ƒ', 'ç„¶', 'æ•…', 'è™½', 'æˆ–', 'æ—¢',
        # ç°ä»£è™šè¯
        'çš„', 'åœ°', 'å¾—', 'äº†', 'ç€', 'è¿‡', 'å•Š', 'å‘€', 'å‘¢', 'å—', 'å§',
        # åœ°åç›¸å…³
        'å¤©æ´¥', 'æ´¥é—¨', 'æ‚è®°', 'å·', 'ç¯‡', 'ç« ', 'èŠ‚',
        # æ ‡ç‚¹ç©ºæ ¼
        ' ', '\n', '\t', '\r', 'ã€€',
        # æ•°å­—
        'ä¸€', 'äºŒ', 'ä¸‰', 'å››', 'äº”', 'å…­', 'ä¸ƒ', 'å…«', 'ä¹', 'å',
        'ç™¾', 'åƒ', 'ä¸‡'
    }

    try:
        with open(filename, 'r', encoding='utf-8') as f:
            for line in f:
                word = line.strip()
                if word:
                    stopwords.add(word)
        stopwords.update(default_stopwords)
        print(f"âœ… å·²åŠ è½½ {len(stopwords)} ä¸ªåœç”¨è¯ï¼ˆå«é»˜è®¤è¯ï¼‰")
    except FileNotFoundError:
        print(f"âš ï¸  æœªæ‰¾åˆ° {filename}ï¼Œä½¿ç”¨é»˜è®¤åœç”¨è¯åˆ—è¡¨")
        stopwords = default_stopwords

    return stopwords

def preprocess_text(text, stopwords):
    """
    æ–‡æœ¬é¢„å¤„ç†ï¼šåˆ†è¯ã€æ¸…æ´—ã€è¿‡æ»¤
    """
    # ä½¿ç”¨jiebaç²¾ç¡®æ¨¡å¼åˆ†è¯
    words = jieba.lcut(text)

    # è¿‡æ»¤åœç”¨è¯å’Œå•å­—è¯
    filtered_words = []
    for word in words:
        word = word.strip()
        if (len(word) > 1 and  # åªä¿ç•™é•¿åº¦å¤§äº1çš„è¯
            word not in stopwords and  # ä¸åœ¨åœç”¨è¯è¡¨ä¸­
            not re.match(r'^[0-9\s\.\,\!\?ã€‚ï¼Œï¼ï¼Ÿã€Šã€‹ã€ã€‘ã€ï¼›ï¼š\-\â€”\â€“\ï¼ˆ\ï¼‰\ã€Œ\ã€\ã€\ã€\ã€”\ã€•]+$', word)):  # éæ ‡ç‚¹ç¬¦å·
            filtered_words.append(word)

    print(f"ğŸ“Š åŸå§‹åˆ†è¯æ•°: {len(words)}, è¿‡æ»¤åè¯æ•°: {len(filtered_words)}")
    return filtered_words

# ========== 4. å…³é”®è¯é¢‘ç»Ÿè®¡å‡½æ•° ==========
def analyze_word_frequency(words, top_n=50):
    """
    ç»Ÿè®¡è¯é¢‘å¹¶è¿”å›å‰Nä¸ªé«˜é¢‘è¯
    """
    word_freq = Counter(words)
    top_words = word_freq.most_common(top_n)

    # åˆ›å»ºDataFrame
    df_top = pd.DataFrame(top_words, columns=['è¯è¯­', 'é¢‘æ¬¡'])

    return df_top, word_freq

def plot_top_words(df_top, n=20, save_path='top_words_bar.png'):
    """
    ç»˜åˆ¶é«˜é¢‘è¯æ¡å½¢å›¾ï¼ˆæ¨ªå‘ï¼‰
    """
    plt.figure(figsize=(14, 10))

    # è®¾ç½®ä¸­æ–‡å­—ä½“
    plt.rcParams['font.sans-serif'] = ['WenQuanYi Micro Hei']
    plt.rcParams['axes.unicode_minus'] = False

    # å–å‰Nä¸ªè¯
    df_plot = df_top.head(n).sort_values('é¢‘æ¬¡', ascending=True)

    # åˆ›å»ºé¢œè‰²æ¸å˜
    colors = plt.cm.Blues(np.linspace(0.6, 0.9, len(df_plot)))

    bars = plt.barh(range(len(df_plot)), df_plot['é¢‘æ¬¡'],
                    color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)

    # æ·»åŠ æ•°å€¼æ ‡ç­¾
    for i, (value, word) in enumerate(zip(df_plot['é¢‘æ¬¡'], df_plot['è¯è¯­'])):
        plt.text(value + max(df_plot['é¢‘æ¬¡'])*0.01, i,
                f'{value}', va='center', fontsize=11, fontweight='bold')

    # è®¾ç½®Yè½´æ ‡ç­¾
    plt.yticks(range(len(df_plot)), df_plot['è¯è¯­'],
              fontsize=13, fontproperties='WenQuanYi Micro Hei')
    plt.xlabel('å‡ºç°é¢‘æ¬¡', fontsize=14, fontweight='bold')
    plt.title(f'ã€Šæ´¥é—¨æ‚è®°ã€‹å‰{n}ä¸ªé«˜é¢‘è¯æ±‡',
             fontsize=18, fontweight='bold', pad=25, fontproperties='WenQuanYi Micro Hei')

    # ç¾åŒ–å›¾è¡¨
    plt.grid(axis='x', alpha=0.3, linestyle='--', linewidth=0.5)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_linewidth(0.5)
    plt.gca().spines['bottom'].set_linewidth(0.5)

    # è°ƒæ•´å¸ƒå±€
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
    plt.show()
    print(f"âœ… é«˜é¢‘è¯æ¡å½¢å›¾å·²ä¿å­˜: {save_path}")

# ========== 5. ä¸»é¢˜è¯è¿½è¸ªå‡½æ•° ==========
def track_theme_words(text, transport_words, commerce_words):
    """
    è¿½è¸ªç‰¹å®šä¸»é¢˜è¯çš„å‡ºç°é¢‘æ¬¡
    """
    # ç»Ÿè®¡ä¸»é¢˜è¯é¢‘æ¬¡
    transport_counts = {}
    commerce_counts = {}

    for word in transport_words:
        count = len(re.findall(word, text))
        transport_counts[word] = count

    for word in commerce_words:
        count = len(re.findall(word, text))
        commerce_counts[word] = count

    # åˆ›å»ºæ±‡æ€»è¡¨æ ¼
    transport_df = pd.DataFrame(list(transport_counts.items()),
                               columns=['äº¤é€šç›¸å…³è¯', 'é¢‘æ¬¡'])
    commerce_df = pd.DataFrame(list(commerce_counts.items()),
                              columns=['å•†ä¸šç›¸å…³è¯', 'é¢‘æ¬¡'])

    # åˆå¹¶è¡¨æ ¼
    theme_df = pd.DataFrame({
        'äº¤é€šç›¸å…³è¯': list(transport_counts.keys()),
        'äº¤é€šé¢‘æ¬¡': list(transport_counts.values()),
        'å•†ä¸šç›¸å…³è¯': list(commerce_counts.keys()),
        'å•†ä¸šé¢‘æ¬¡': list(commerce_counts.values())
    })

    return theme_df, transport_counts, commerce_counts

def analyze_chapter_distribution(text, transport_words, commerce_words):
    """
    åˆ†æä¸»é¢˜è¯åœ¨ä¸åŒç« èŠ‚çš„åˆ†å¸ƒ
    """
    # å°è¯•æŒ‰ç« èŠ‚åˆ†å‰²æ–‡æœ¬
    chapter_patterns = [
        r'å·[ä¸€äºŒä¸‰å››äº”å…­ä¸ƒå…«ä¹å]+',
        r'ç¬¬[ä¸€äºŒä¸‰å››äº”å…­ä¸ƒå…«ä¹å]+å·',
        r'ç¬¬[ä¸€äºŒä¸‰å››äº”å…­ä¸ƒå…«ä¹å]+ç« ',
        r'[ä¸€äºŒä¸‰å››äº”å…­ä¸ƒå…«ä¹å]+ã€',
        r'å·\d+',
        r'ç¬¬\d+å·',
        r'ç¬¬\d+ç« '
    ]

    # æŸ¥æ‰¾æ‰€æœ‰ç« èŠ‚æ ‡é¢˜ä½ç½®
    chapter_positions = []
    for pattern in chapter_patterns:
        matches = list(re.finditer(pattern, text))
        for match in matches:
            # æ£€æŸ¥æ˜¯å¦æ˜¯æœ‰æ•ˆçš„ç« èŠ‚æ ‡é¢˜ï¼ˆåé¢æœ‰æ¢è¡Œæˆ–å†…å®¹ï¼‰
            pos = match.end()
            if pos < len(text) and text[pos] in ['\n', 'ï¼Œ', 'ã€‚', 'ï¼š', ' ']:
                chapter_positions.append(match)

    if len(chapter_positions) < 2:
        print("âš ï¸  æœªæ£€æµ‹åˆ°æ˜æ˜¾çš„ç« èŠ‚ç»“æ„ï¼Œå°†è¿›è¡Œå…¨æ–‡åˆ†æ")
        return None

    # æŒ‰ä½ç½®æ’åº
    chapter_positions.sort(key=lambda x: x.start())

    # åˆ†å‰²ç« èŠ‚
    chapters_data = []
    for i in range(len(chapter_positions)):
        start = chapter_positions[i].start()
        if i < len(chapter_positions) - 1:
            end = chapter_positions[i+1].start()
        else:
            end = len(text)

        chapter_title = chapter_positions[i].group()
        chapter_content = text[start:end]
        chapters_data.append((chapter_title, chapter_content))

    print(f"ğŸ“š æ£€æµ‹åˆ° {len(chapters_data)} ä¸ªç« èŠ‚")

    # ç»Ÿè®¡æ¯ä¸ªç« èŠ‚çš„ä¸»é¢˜è¯é¢‘æ¬¡
    chapter_stats = []
    for title, content in chapters_data:
        stats = {'ç« èŠ‚': title, 'å†…å®¹é•¿åº¦': len(content)}

        # ç»Ÿè®¡äº¤é€šç›¸å…³è¯
        for word in transport_words:
            stats[f'äº¤é€š_{word}'] = len(re.findall(word, content))

        # ç»Ÿè®¡å•†ä¸šç›¸å…³è¯
        for word in commerce_words:
            stats[f'å•†ä¸š_{word}'] = len(re.findall(word, content))

        chapter_stats.append(stats)

    chapter_df = pd.DataFrame(chapter_stats)

    # ç»˜åˆ¶ç« èŠ‚åˆ†å¸ƒå›¾
    if len(chapters_data) > 1:
        plot_chapter_distribution(chapter_df, transport_words, commerce_words)

    return chapter_df

def plot_chapter_distribution(chapter_df, transport_words, commerce_words):
    """
    ç»˜åˆ¶ä¸»é¢˜è¯ç« èŠ‚åˆ†å¸ƒå›¾
    """
    plt.rcParams['font.sans-serif'] = ['WenQuanYi Micro Hei']
    plt.rcParams['axes.unicode_minus'] = False

    # æå–äº¤é€šç›¸å…³è¯æ•°æ®
    transport_cols = [col for col in chapter_df.columns if col.startswith('äº¤é€š_')]
    commerce_cols = [col for col in chapter_df.columns if col.startswith('å•†ä¸š_')]

    if len(transport_cols) > 0:
        fig, axes = plt.subplots(1, 2, figsize=(16, 6))

        # äº¤é€šè¯åˆ†å¸ƒ
        ax1 = axes[0]
        transport_data = chapter_df[transport_cols]
        transport_data.index = chapter_df['ç« èŠ‚'].apply(lambda x: x[:10] + '...' if len(x) > 10 else x)

        # åˆ›å»ºå †å æŸ±çŠ¶å›¾
        bottom = np.zeros(len(transport_data))
        colors = plt.cm.Set3(np.linspace(0, 1, len(transport_cols)))

        for i, col in enumerate(transport_cols):
            word = col.replace('äº¤é€š_', '')
            values = transport_data[col].values
            ax1.bar(transport_data.index, values, bottom=bottom,
                   label=word, color=colors[i], alpha=0.8)
            bottom += values

        ax1.set_title('äº¤é€šç›¸å…³è¯ç« èŠ‚åˆ†å¸ƒ', fontsize=16, fontweight='bold', pad=15)
        ax1.set_xlabel('ç« èŠ‚', fontsize=12)
        ax1.set_ylabel('å‡ºç°é¢‘æ¬¡', fontsize=12)
        ax1.tick_params(axis='x', rotation=45)
        ax1.legend(title='äº¤é€šè¯è¯­', bbox_to_anchor=(1.05, 1), loc='upper left')
        ax1.grid(axis='y', alpha=0.3, linestyle='--')

        # å•†ä¸šè¯åˆ†å¸ƒ
        if len(commerce_cols) > 0:
            ax2 = axes[1]
            commerce_data = chapter_df[commerce_cols]
            commerce_data.index = chapter_df['ç« èŠ‚'].apply(lambda x: x[:10] + '...' if len(x) > 10 else x)

            bottom = np.zeros(len(commerce_data))
            colors = plt.cm.Set2(np.linspace(0, 1, len(commerce_cols)))

            for i, col in enumerate(commerce_cols):
                word = col.replace('å•†ä¸š_', '')
                values = commerce_data[col].values
                ax2.bar(commerce_data.index, values, bottom=bottom,
                       label=word, color=colors[i], alpha=0.8)
                bottom += values

            ax2.set_title('å•†ä¸šç›¸å…³è¯ç« èŠ‚åˆ†å¸ƒ', fontsize=16, fontweight='bold', pad=15)
            ax2.set_xlabel('ç« èŠ‚', fontsize=12)
            ax2.set_ylabel('å‡ºç°é¢‘æ¬¡', fontsize=12)
            ax2.tick_params(axis='x', rotation=45)
            ax2.legend(title='å•†ä¸šè¯è¯­', bbox_to_anchor=(1.05, 1), loc='upper left')
            ax2.grid(axis='y', alpha=0.3, linestyle='--')

        plt.tight_layout()
        plt.savefig('chapter_distribution.png', dpi=300, bbox_inches='tight', facecolor='white')
        plt.show()
        print("âœ… ç« èŠ‚åˆ†å¸ƒå›¾å·²ä¿å­˜")

def plot_theme_words_comparison(transport_counts, commerce_counts):
    """
    ç»˜åˆ¶ä¸»é¢˜è¯å¯¹æ¯”å›¾
    """
    plt.rcParams['font.sans-serif'] = ['WenQuanYi Micro Hei']
    plt.rcParams['axes.unicode_minus'] = False

    fig, axes = plt.subplots(1, 2, figsize=(16, 7))

    # äº¤é€šç›¸å…³è¯
    ax1 = axes[0]
    transport_items = sorted(transport_counts.items(), key=lambda x: x[1], reverse=True)
    transport_words_sorted = [item[0] for item in transport_items]
    transport_counts_sorted = [item[1] for item in transport_items]

    colors1 = plt.cm.Reds(np.linspace(0.6, 0.9, len(transport_words_sorted)))
    bars1 = ax1.barh(transport_words_sorted, transport_counts_sorted, color=colors1, alpha=0.8)

    ax1.set_xlabel('å‡ºç°é¢‘æ¬¡', fontsize=12, fontweight='bold')
    ax1.set_title('äº¤é€šç›¸å…³è¯é¢‘æ¬¡ç»Ÿè®¡', fontsize=15, fontweight='bold', pad=15)
    ax1.invert_yaxis()
    ax1.grid(axis='x', alpha=0.3, linestyle='--')

    # æ·»åŠ æ•°å€¼æ ‡ç­¾
    for i, v in enumerate(transport_counts_sorted):
        ax1.text(v + max(transport_counts_sorted)*0.02, i, str(v),
                va='center', fontsize=11, fontweight='bold')

    # å•†ä¸šç›¸å…³è¯
    ax2 = axes[1]
    commerce_items = sorted(commerce_counts.items(), key=lambda x: x[1], reverse=True)
    commerce_words_sorted = [item[0] for item in commerce_items]
    commerce_counts_sorted = [item[1] for item in commerce_items]

    colors2 = plt.cm.Greens(np.linspace(0.6, 0.9, len(commerce_words_sorted)))
    bars2 = ax2.barh(commerce_words_sorted, commerce_counts_sorted, color=colors2, alpha=0.8)

    ax2.set_xlabel('å‡ºç°é¢‘æ¬¡', fontsize=12, fontweight='bold')
    ax2.set_title('å•†ä¸šç›¸å…³è¯é¢‘æ¬¡ç»Ÿè®¡', fontsize=15, fontweight='bold', pad=15)
    ax2.invert_yaxis()
    ax2.grid(axis='x', alpha=0.3, linestyle='--')

    # æ·»åŠ æ•°å€¼æ ‡ç­¾
    for i, v in enumerate(commerce_counts_sorted):
        ax2.text(v + max(commerce_counts_sorted)*0.02, i, str(v),
                va='center', fontsize=11, fontweight='bold')

    plt.tight_layout()
    plt.savefig('theme_words_distribution.png', dpi=300, bbox_inches='tight', facecolor='white')
    plt.show()
    print("âœ… ä¸»é¢˜è¯åˆ†å¸ƒå›¾å·²ä¿å­˜")

# ========== 6. ä¸Šä¸‹æ–‡è¯­ä¹‰åˆ†æå‡½æ•° ==========
def extract_keyword_contexts(text, keyword, context_chars=150):
    """
    æå–å…³é”®è¯çš„ä¸Šä¸‹æ–‡è¯­å¢ƒ
    """
    # æŸ¥æ‰¾æ‰€æœ‰å…³é”®è¯å‡ºç°çš„ä½ç½®
    contexts = []
    pattern = re.compile(keyword)

    for match in pattern.finditer(text):
        start = max(0, match.start() - context_chars)
        end = min(len(text), match.end() + context_chars)

        context = text[start:end]
        # æ¸…ç†æ¢è¡Œå’Œå¤šä½™ç©ºæ ¼
        context = re.sub(r'\s+', ' ', context).strip()
        contexts.append(context)

    return contexts

# ========== 7. è¯äº‘ç”Ÿæˆå‡½æ•° ==========
def generate_wordcloud(words, word_freq, save_path='wordcloud.png'):
    """
    ç”Ÿæˆè¯äº‘å›¾
    """
    try:
        # å°†è¯é¢‘è½¬æ¢ä¸ºå­—å…¸
        freq_dict = dict(word_freq)

        # åˆ›å»ºè¯äº‘å¯¹è±¡
        wc = WordCloud(
            font_path='/usr/share/fonts/truetype/wqy/wqy-microhei.ttc',
            width=1200,
            height=800,
            background_color='white',
            max_words=100,
            max_font_size=150,
            random_state=42,
            colormap='viridis'
        )

        # ç”Ÿæˆè¯äº‘
        wc.generate_from_frequencies(freq_dict)

        # ç»˜åˆ¶è¯äº‘
        plt.figure(figsize=(15, 10))
        plt.imshow(wc, interpolation='bilinear')
        plt.axis('off')
        plt.title('ã€Šæ´¥é—¨æ‚è®°ã€‹è¯äº‘å›¾', fontsize=20, fontweight='bold', pad=20)
        plt.tight_layout()
        plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        plt.show()
        print(f"âœ… è¯äº‘å›¾å·²ä¿å­˜: {save_path}")

    except Exception as e:
        print(f"âš ï¸  ç”Ÿæˆè¯äº‘æ—¶å‡ºé”™: {e}")

# ========== 8. ä¸»åˆ†ææµç¨‹ ==========
def main_analysis():
    """
    ä¸»åˆ†æå‡½æ•°
    """
    print("=" * 70)
    print("ğŸ“– ã€Šæ´¥é—¨æ‚è®°ã€‹æ–‡æœ¬åˆ†æ - å¤©æ´¥å•†ä¸šç©ºé—´ä¸äº¤é€šæ–¹å¼å˜è¿ç ”ç©¶")
    print("=" * 70)

    # 1. è¯»å–æ–‡æœ¬
    print("\n1ï¸âƒ£ è¯»å–æ–‡æœ¬æ–‡ä»¶...")
    try:
        text = read_text_file('jinmen_zaji.txt')
        print(f"ğŸ“„ æ–‡æœ¬é•¿åº¦: {len(text):,} å­—ç¬¦")
    except Exception as e:
        print(f"âŒ è¯»å–æ–‡ä»¶å¤±è´¥: {e}")
        print("è¯·ç¡®ä¿æ–‡ä»¶ 'jinmen_zaji.txt' å­˜åœ¨")
        return

    # 2. åŠ è½½åœç”¨è¯
    print("\n2ï¸âƒ£ åŠ è½½åœç”¨è¯è¡¨...")
    stopwords = load_stopwords('stopwords.txt')

    # 3. æ–‡æœ¬é¢„å¤„ç†
    print("\n3ï¸âƒ£ æ–‡æœ¬é¢„å¤„ç†...")
    words = preprocess_text(text, stopwords)

    # 4. å…³é”®è¯é¢‘åˆ†æ
    print("\n4ï¸âƒ£ å…³é”®è¯é¢‘åˆ†æ...")
    df_top, word_freq = analyze_word_frequency(words, top_n=50)

    # ä¿å­˜é«˜é¢‘è¯
    df_top.to_csv('top_keywords.csv', index=False, encoding='utf-8-sig')
    print(f"âœ… é«˜é¢‘è¯å·²ä¿å­˜è‡³: top_keywords.csv")
    print(f"ğŸ“Š å‰10ä¸ªé«˜é¢‘è¯:")
    print(df_top.head(10).to_string(index=False))

    # ç»˜åˆ¶é«˜é¢‘è¯æ¡å½¢å›¾
    plot_top_words(df_top, n=20)

    # 5. ç”Ÿæˆè¯äº‘å›¾
    print("\n5ï¸âƒ£ ç”Ÿæˆè¯äº‘å›¾...")
    generate_wordcloud(words, word_freq)

    # 6. ä¸»é¢˜è¯è¿½è¸ª
    print("\n6ï¸âƒ£ ä¸»é¢˜è¯è¿½è¸ªåˆ†æ...")

    # å®šä¹‰ä¸»é¢˜è¯é›†
    transport_words = ["é“è·¯", "ç«è½¦", "è½¦ç«™", "ç å¤´", "è¿æ²³", "èˆŸè½¦", "ç”µè½¦", "é©¬è·¯"]
    commerce_words = ["å¸‚åœº", "å•†é“º", "é“¶è¡Œ", "å…¬å¸", "åŠä¸šåœº", "ç¹å", "è´¸æ˜“", "é“¶å·"]

    # ç»Ÿè®¡ä¸»é¢˜è¯é¢‘æ¬¡
    theme_df, transport_counts, commerce_counts = track_theme_words(
        text, transport_words, commerce_words)

    # ä¿å­˜ä¸»é¢˜è¯ç»Ÿè®¡
    theme_df.to_csv('theme_keywords.csv', index=False, encoding='utf-8-sig')
    print(f"âœ… ä¸»é¢˜è¯ç»Ÿè®¡å·²ä¿å­˜è‡³: theme_keywords.csv")

    # æ‰“å°ä¸»é¢˜è¯ç»Ÿè®¡æ‘˜è¦
    print("\nğŸ“ˆ ä¸»é¢˜è¯ç»Ÿè®¡æ‘˜è¦:")
    print("-" * 40)
    print("äº¤é€šç›¸å…³è¯:")
    for word, count in sorted(transport_counts.items(), key=lambda x: x[1], reverse=True):
        print(f"  {word}: {count}æ¬¡")

    print("\nå•†ä¸šç›¸å…³è¯:")
    for word, count in sorted(commerce_counts.items(), key=lambda x: x[1], reverse=True):
        print(f"  {word}: {count}æ¬¡")
    print("-" * 40)

    # ç»˜åˆ¶ä¸»é¢˜è¯å¯¹æ¯”å›¾
    plot_theme_words_comparison(transport_counts, commerce_counts)

    # 7. ç« èŠ‚åˆ†å¸ƒåˆ†æ
    print("\n7ï¸âƒ£ ç« èŠ‚åˆ†å¸ƒåˆ†æ...")
    chapter_df = analyze_chapter_distribution(text, transport_words, commerce_words)
    if chapter_df is not None:
        chapter_df.to_csv('chapter_analysis.csv', index=False, encoding='utf-8-sig')
        print(f"âœ… ç« èŠ‚åˆ†æç»“æœå·²ä¿å­˜è‡³: chapter_analysis.csv")

        # æ‰“å°ç« èŠ‚åˆ†å¸ƒæ‘˜è¦
        if 'ç« èŠ‚' in chapter_df.columns:
            print("\nğŸ“‘ ç« èŠ‚åˆ†å¸ƒæ‘˜è¦:")
            print("-" * 50)
            for idx, row in chapter_df.iterrows():
                transport_total = sum([row.get(f'äº¤é€š_{word}', 0) for word in transport_words])
                commerce_total = sum([row.get(f'å•†ä¸š_{word}', 0) for word in commerce_words])
                print(f"{row['ç« èŠ‚']}: äº¤é€šè¯{transport_total}æ¬¡, å•†ä¸šè¯{commerce_total}æ¬¡")
            print("-" * 50)

    # 8. ä¸Šä¸‹æ–‡è¯­ä¹‰åˆ†æ
    print("\n8ï¸âƒ£ ä¸Šä¸‹æ–‡è¯­ä¹‰åˆ†æ...")
    keyword = "é“è·¯"
    contexts = extract_keyword_contexts(text, keyword, context_chars=150)

    # ä¿å­˜ä¸Šä¸‹æ–‡
    with open('railway_contexts.txt', 'w', encoding='utf-8') as f:
        f.write(f"å…³é”®è¯ '{keyword}' çš„ä¸Šä¸‹æ–‡åˆ†æ\n")
        f.write(f"å…±æ‰¾åˆ° {len(contexts)} å¤„å‡ºç°\n")
        f.write("=" * 70 + "\n\n")

        for i, context in enumerate(contexts, 1):
            f.write(f"ã€ç¬¬{i}å¤„ã€‘\n")
            f.write(f"{context}\n")
            f.write("-" * 70 + "\n\n")

    print(f"âœ… å…³é”®è¯ '{keyword}' çš„ä¸Šä¸‹æ–‡å·²ä¿å­˜è‡³: railway_contexts.txt")
    print(f"ğŸ” å…±æ‰¾åˆ° {len(contexts)} å¤„å‡ºç°")

    # æ‰“å°éƒ¨åˆ†ä¸Šä¸‹æ–‡ç¤ºä¾‹
    if contexts:
        print("\nğŸ“ ä¸Šä¸‹æ–‡ç¤ºä¾‹ï¼ˆå‰3å¤„ï¼‰:")
        print("-" * 60)
        for i, context in enumerate(contexts[:3], 1):
            print(f"ç¤ºä¾‹{i}: {context[:100]}...")
        print("-" * 60)

    # 9. æ±‡æ€»ç»Ÿè®¡
    print("\n" + "=" * 70)
    print("ğŸ‰ åˆ†æå®Œæˆï¼æ±‡æ€»ç»“æœï¼š")
    print("=" * 70)
    print(f"ğŸ“Š åŸºæœ¬ç»Ÿè®¡:")
    print(f"  â€¢ æ–‡æœ¬æ€»å­—ç¬¦æ•°: {len(text):,}")
    print(f"  â€¢ æœ‰æ•ˆè¯æ±‡æ€»æ•°: {len(words):,}")
    print(f"  â€¢ å”¯ä¸€è¯æ±‡æ•°é‡: {len(word_freq):,}")

    print(f"\nğŸ“ ç”Ÿæˆæ–‡ä»¶:")
    print(f"  1. é«˜é¢‘è¯CSV: top_keywords.csv")
    print(f"  2. ä¸»é¢˜è¯CSV: theme_keywords.csv")
    print(f"  3. ä¸Šä¸‹æ–‡æ–‡æœ¬: railway_contexts.txt")
    print(f"  4. é«˜é¢‘è¯æ¡å½¢å›¾: top_words_bar.png")
    print(f"  5. ä¸»é¢˜è¯åˆ†å¸ƒå›¾: theme_words_distribution.png")
    print(f"  6. è¯äº‘å›¾: wordcloud.png")

    if chapter_df is not None:
        print(f"  7. ç« èŠ‚åˆ†æCSV: chapter_analysis.csv")
        print(f"  8. ç« èŠ‚åˆ†å¸ƒå›¾: chapter_distribution.png")

    # ç»Ÿè®¡æœ€ç›¸å…³çš„10ä¸ªè¯
    print(f"\nğŸ”‘ é‡è¦å‘ç°:")

    # æ‰¾å‡ºä¸äº¤é€šå•†ä¸šç›¸å…³çš„å…¶ä»–é«˜é¢‘è¯
    related_words = []
    for word, freq in word_freq.most_common(100):
        if (any(t in word for t in transport_words + commerce_words) or
            len(word) >= 3):  # é•¿è¯é€šå¸¸æ›´æœ‰æ„ä¹‰
            if word not in transport_words + commerce_words:
                related_words.append((word, freq))

    if related_words:
        print(f"  å…¶ä»–ç›¸å…³é«˜é¢‘è¯ï¼ˆå‰10ï¼‰:")
        for word, freq in related_words[:10]:
            print(f"    {word}: {freq}æ¬¡")

    print("=" * 70)

# ========== 9. åˆ›å»ºç¤ºä¾‹åœç”¨è¯æ–‡ä»¶ ==========
def create_sample_stopwords():
    """
    åˆ›å»ºç¤ºä¾‹åœç”¨è¯æ–‡ä»¶
    """
    stopwords_content = """# åœç”¨è¯è¡¨ - ã€Šæ´¥é—¨æ‚è®°ã€‹æ–‡æœ¬åˆ†æ
# æ–‡è¨€è™šè¯
ä¹‹
ä¹
è€…
ä¹Ÿ
çŸ£
ç„‰
å“‰
æ›°
äº‘
è€Œ
ä»¥
å…¶
æ‰€
ä¸º
äº
åˆ™
ä½†
çš†
äº¦
åˆ
æœ‰
æ— 
æœª
é
ä¸
ä¸”
ä¹ƒ
ç„¶
æ•…
è™½
æˆ–
æ—¢

# ç°ä»£è™šè¯
çš„
åœ°
å¾—
äº†
ç€
è¿‡
å•Š
å‘€
å‘¢
å—
å§

# åœ°åç›¸å…³
å¤©æ´¥
æ´¥é—¨
æ‚è®°
å·
ç¯‡
ç« 
èŠ‚

# äººç§°ä»£è¯
ä½™
å¾
æˆ‘
å°”
æ±
å½¼
å…¶
ä¹‹

# å…¶ä»–å¸¸è§è¯
æ˜¯
æ­¤
å½¼
æ­¤
å³
ä¾¿
åˆ™
è™½
ä»
å¹¶
æ›¾"""

    with open('stopwords.txt', 'w', encoding='utf-8') as f:
        f.write(stopwords_content)
    print("âœ… å·²åˆ›å»ºç¤ºä¾‹åœç”¨è¯æ–‡ä»¶: stopwords.txt")

# ========== æ‰§è¡Œä¸»å‡½æ•° ==========
if __name__ == "__main__":
    # æ£€æŸ¥å¿…è¦åº“
    print("ğŸ” æ£€æŸ¥ç¯å¢ƒé…ç½®...")
    try:
        import jieba
        import pandas as pd
        import matplotlib
        print("âœ… æ‰€æœ‰å¿…è¦åº“å·²å¯¼å…¥æˆåŠŸ")
    except ImportError as e:
        print(f"âŒ å¯¼å…¥åº“å¤±è´¥: {e}")
        print("è¯·ç¡®ä¿å·²å®‰è£…ä»¥ä¸‹åº“: jieba, pandas, matplotlib, seaborn, wordcloud")
        print("åœ¨Colabä¸­è¿è¡Œä»¥ä¸‹å‘½ä»¤å®‰è£…:")
        print("!pip install jieba pandas matplotlib seaborn wordcloud")

    # åˆ›å»ºç¤ºä¾‹åœç”¨è¯æ–‡ä»¶ï¼ˆå¦‚æœä¸å­˜åœ¨ï¼‰
    if not os.path.exists('stopwords.txt'):
        print("\nğŸ“ åˆ›å»ºåœç”¨è¯æ–‡ä»¶...")
        create_sample_stopwords()

    # è¿è¡Œåˆ†æ
    main_analysis()

    print("\nğŸ¯ åˆ†ææµç¨‹å®Œæˆï¼")
    print("ğŸ“Œ æç¤ºï¼šæ‰€æœ‰å›¾è¡¨å’Œæ–‡ä»¶å·²ä¿å­˜åœ¨å½“å‰ç›®å½•")
```
